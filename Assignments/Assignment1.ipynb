```
import pandas as pd

bank_note_data = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/bank_note_data/training_set_label.csv" )
```

```
test_data = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/bank_note_data/testing_set_label.csv')

```

```
bank_note_data.head()
```

```
test_data.head()
```

```
import numpy as np
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
```

```
X_train = bank_note_data.drop("Class",axis=1)
y_train = bank_note_data["Class"]
print(X_train.shape)
print(y_train.shape)
print(X_train.shape[1])
```

```
model = Sequential()
model.add(Dense(32, activation="relu", input_shape=(X_train.shape[1],) ))
model.add(Dense(16, activation="relu"))
model.add(Dense(1, activation="sigmoid"))
model.summary()
```

```
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```
model.fit(X_train, y_train, epochs=15, batch_size=40, verbose=1, validation_split=0.2)
```

```
test_data.shape
```

```
y = model.predict(test_data)
y
```

```
op = [1 if x>0.5 else 0 for x in y ]
```

```
op
```

```
# To create Dataframe of predicted value with particular respective index
res = pd.DataFrame(op) #preditcions are nothing but the final predictions of your model on input features of your new unseen test data
res.index = test_data.index # its important for comparison. Here "test_new" is your new test dataset
res.columns = ["prediction"]

# To download the csv file locally
from google.colab import files
res.to_csv('prediction_results.csv')         
files.download('prediction_results.csv')
```
